```python
from megnet.utils.models import load_model, AVAILABLE_MODELS
from pymatgen.core import Structure, Lattice
import pandas as pd
import numpy as np
import os
from tqdm import tqdm

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)


```

```python
feature = pd.read_csv('data/Data_S1.csv')
MOFList = np.unique(feature['MOF'])

df = pd.DataFrame()

cifFile_dir = '/home/zhenzi/Documents/CORE/CoRE-MOF-1.0-DFT-minimized/CoRE-MOF-1.0-DFT-Minimized/minimized_structures_with_DDEC_charges'
cifFile_dir2 = '/home/zhenzi/Documents/CORE/CoRE-MOF-1.0-DFT-minimized/CoRE-MOF-1.0-DFT-Minimized/minimized_structures'

def model_predict(name,structure):
    new_row = {'MOF': name}
    for model_name in AVAILABLE_MODELS:
        if model_name.endswith("_2019"):
            model = load_model(model_name)
            model.metadata
        
            if model_name.startswith("log"):
                prediction = 10 ** model.predict_structure(structure).ravel()[0]
            else:
                prediction = model.predict_structure(structure).ravel()[0]
            prop_name = model_name.split("_")[0].replace("log", "")
            new_row[prop_name] = prediction
            # new_row['unit'] = model.metadata["unit"].replace("log", "")

    return new_row

for MOF in tqdm(MOFList):
    try:
        MOF_cif = MOF + '_clean_min_charges.cif'
        structure = Structure.from_file( os.path.join(cifFile_dir,MOF_cif) )
    except:
        try:
            MOF_cif = MOF + '_clean_min.cif'
            structure = Structure.from_file( os.path.join(cifFile_dir2,MOF_cif) )
        except:
            continue
        
    new_row = model_predict(MOF,structure)
    df = df.append(new_row, ignore_index=True)
    

df.to_csv('megnet_properties.csv')
```
